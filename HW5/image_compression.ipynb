```python
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread

%matplotlib inline 
```

# Part a)

```python
def low_rank_approximation(X, rank):
    # YOUR CODE GOES HERE
    s,v,d = np.linalg.svd(X,full_matrices=False)
    v[rank:] = 0
    return s.dot(np.diag(v)).dot(d)
```

```python
face = imread("./face.jpg")
```

```python
face_img = [ind[:, 0] for ind in face]
```

```python
face_rank5 = low_rank_approximation(face_img, int(512 * 0.01))
face_rank20 = low_rank_approximation(face_img, int(512 * 0.1))
face_rank100 = low_rank_approximation(face_img, int(512 * 0.5))
```

```python
plt.imshow(face_rank5,cmap='gray')
plt.title('1%')
#plt.savefig('face_rank5.jpg')
```

```python
plt.imshow(face_rank20, cmap='gray')
plt.title('10%')
#plt.savefig('face_rank20.jpg')
```

```python
plt.imshow(face_rank100,cmap='gray')
plt.title('50%')
#plt.savefig('face_rank100.jpg')
```
